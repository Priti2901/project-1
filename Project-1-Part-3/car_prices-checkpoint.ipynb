```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path

car_prices_to_load=Path("car_prices.csv")
car_prices=pd.read_csv(car_prices_to_load)
car_prices.head()
```

```python
car_prices.count()
```

```python
car_prices_new=car_prices.dropna(how='any')
car_prices_new.count()
```

```python
cp_states_1=car_prices_new.groupby(["state"])["make"].count()
cp_states_1.head()
```

```python
x_axis=np.arange(len(cp_states_1))
tick_locations=[value for value in x_axis]
plt.bar(x_axis,cp_states_1,color='r',alpha=0.5,align="center")
plt.xticks(tick_locations,cp_states_1,rotation="vertical")
plt.show()

```
